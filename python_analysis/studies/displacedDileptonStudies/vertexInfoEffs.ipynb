```python
import sys
import re
sys.path.append("/uscms/home/sbrightt/nobackup/iDM/iDMe_analysis/CMSSW_10_6_26/src/iDMeAnalysis/python_analysis/analysisTools")
import plotTools as ptools
import matplotlib.pyplot as plt
import numpy as np
from coffea import hist
import importlib
import os
from analysisTools import fileAnalyzer, Analyzer

ptools.setDefaultStyle()
#plt.style.use(cfgDir+'plt_settings.mplstyle')
```

```python
histoLoc = "/uscms_data/d3/sbrightt/iDM/iDMe_analysis/CMSSW_10_6_26/src/iDMeAnalysis/python_analysis/outputs/signal_2018_aEM"
histos = ptools.loadHistoFiles(histoLoc)
info, masses, cts = ptools.getSampleInfo(histos)
```

```python
# plotting vertex information (quality, kinematics, etc)
h_kin = histos["dispEE_vtx_kinematics"]
h_qual = histos["dispEE_vtx_qual"]
vxy_top = {"1":5,"10":15,"100":20,"1000":20}
importlib.reload(ptools)
ptools.setDefaultStyle()

for pt in list(info.keys()):
    for ct in list(info[pt].keys()):
        m, dm = pt.split("-")[0], pt.split("-")[1]
        samp = info[pt][ct]
        outdir = 'plots/individuals/dispEE_info/{0}-{1}/ct-{2}/'.format(m,dm,ct)
        os.makedirs(outdir,exist_ok=True)
        title = ptools.reduceSampleName(samp,full=True)
        vxy_int = [0,vxy_top[ct]]
        
        hsamp_kin = h_kin.integrate("sample",int_range=samp)
        hsamp_qual = h_qual.integrate("sample",int_range=samp)
        
        plt.figure()
        h_mass = hsamp_kin.sum("leadPt","dR",overflow="all")
        hist.plot1d(h_mass,overlay="denom")
        plt.title(title)
        plt.gca().get_legend().remove()
        plt.savefig(outdir+"dispEE_mass.pdf")
        plt.close()
        
        plt.figure()
        h_leadPt = hsamp_kin.sum("mass","dR",overflow="all")
        hist.plot1d(h_leadPt,overlay="denom")
        plt.title(title)
        plt.gca().get_legend().remove()
        plt.savefig(outdir+"dispEE_leadPt.pdf")
        plt.close()
        
        plt.figure()
        h_dR = hsamp_kin.sum("mass","leadPt",overflow="all")
        hist.plot1d(h_dR,overlay="denom")
        plt.title(title)
        plt.gca().get_legend().remove()
        plt.savefig(outdir+"dispEE_dR.pdf")
        plt.close()
        
        plt.figure()
        h_vxy = hsamp_qual.sum("dxy","chi2",overflow="all")
        plt.yscale("log")
        hist.plot1d(h_vxy,overlay="denom")
        plt.xlim(vxy_int)
        plt.title(title)
        plt.gca().get_legend().remove()
        plt.savefig(outdir+"dispEE_vxy.pdf")
        plt.close()
        
        plt.figure()
        h_dxy = hsamp_qual.sum("vxy","chi2",overflow="all")
        plt.yscale("log")
        hist.plot1d(h_dxy,overlay="denom")
        plt.xlim(vxy_int)
        plt.title(title)
        plt.gca().get_legend().remove()
        plt.savefig(outdir+"dispEE_dxy.pdf")
        plt.close()
        
        plt.figure()
        h_chi2 = hsamp_qual.sum("vxy","dxy",overflow="all")
        plt.yscale("log")
        hist.plot1d(h_chi2,overlay="denom")
        plt.title(title)
        plt.gca().get_legend().remove()
        plt.savefig(outdir+"dispEE_chi2.pdf")
        plt.close()
        
# plotting vertex finding efficiency
h_eff = histos["dispEE_eff"]
    
for pt in list(info.keys()):
    for ct in list(info[pt].keys()):
        m, dm = pt.split("-")[0], pt.split("-")[1]
        samp = info[pt][ct]
        outdir = 'plots/individuals/dispEE_eff/{0}-{1}/ct-{2}/'.format(m,dm,ct)
        os.makedirs(outdir,exist_ok=True)
        title = ptools.reduceSampleName(samp,full=True)
        
        plt.figure()
        hsamp_eff = h_eff.integrate("sample",int_range=samp)
        hist.plot1d(hsamp_eff,overlay="denom",density=True)
        plt.title(title)
        plt.gca().get_legend().remove()
        plt.savefig(outdir+"dispEE_eff_byDenom.pdf")
        plt.close()
```

```python
# Same plots as above, with ctau overlay
# split up by the denom categories
h_kin = histos["dispEE_vtx_kinematics"]
h_qual = histos["dispEE_vtx_qual"]
vxy_top = {"1":5,"10":15,"100":20,"1000":20}
importlib.reload(ptools)
ptools.setDefaultStyle()
denoms = ["all","noRegVtx","noGenMatchEles"]

for pt in masses:
    for denom in denoms:
        m, dm = pt.split("-")[0], pt.split("-")[1]
        outdir = 'plots/ct-overlays/dispEE_info/denom-{0}/{1}-{2}/'.format(denom,m,dm)
        os.makedirs(outdir,exist_ok=True)
        title = "$m_\chi = {0}$ GeV, $\Delta m_\chi = {1}$ GeV".format(m,dm)
        title += ", Denom: {0}".format(denom)
        samps_ct = [info[pt][ct] for ct in list(info[pt].keys())]
        
        hsamp_kin = h_kin.integrate("denom",int_range=denom)
        hsamp_qual = h_qual.integrate("denom",int_range=denom).rebin("vxy",2).rebin("dxy",2)
        
        plt.figure()
        h_mass = hsamp_kin.sum("leadPt","dR",overflow="all")
        hist.plot1d(h_mass,overlay="sample",order=samps_ct,density=True)
        plt.title(title)
        leg = plt.gca().get_legend()
        for txt in leg.get_texts():
            text = txt.get_text()
            txt.set_text(ptools.reduceSampleName(text,lifetime=True))
        plt.savefig(outdir+"dispEE_mass.pdf")
        plt.close()
        
        plt.figure()
        h_leadPt = hsamp_kin.sum("mass","dR",overflow="all")
        hist.plot1d(h_leadPt,overlay="sample",order=samps_ct,density=True)
        plt.title(title)
        leg = plt.gca().get_legend()
        for txt in leg.get_texts():
            text = txt.get_text()
            txt.set_text(ptools.reduceSampleName(text,lifetime=True))
        plt.savefig(outdir+"dispEE_leadPt.pdf")
        plt.close()
        
        plt.figure()
        h_dR = hsamp_kin.sum("mass","leadPt",overflow="all")
        hist.plot1d(h_dR,overlay="sample",order=samps_ct,density=True)
        plt.title(title)
        leg = plt.gca().get_legend()
        for txt in leg.get_texts():
            text = txt.get_text()
            txt.set_text(ptools.reduceSampleName(text,lifetime=True))
        plt.savefig(outdir+"dispEE_dR.pdf")
        plt.close()
        
        plt.figure()
        h_vxy = hsamp_qual.sum("dxy","chi2",overflow="all")
        plt.yscale("log")
        hist.plot1d(h_vxy,overlay="sample",order=samps_ct,density=True)
        plt.xlim([0,20])
        plt.title(title)
        leg = plt.gca().get_legend()
        for txt in leg.get_texts():
            text = txt.get_text()
            txt.set_text(ptools.reduceSampleName(text,lifetime=True))
        plt.savefig(outdir+"dispEE_vxy.pdf")
        plt.close()
        
        plt.figure()
        h_dxy = hsamp_qual.sum("vxy","chi2",overflow="all")
        plt.yscale("log")
        hist.plot1d(h_dxy,overlay="sample",order=samps_ct,density=True)
        plt.xlim([0,20])
        plt.title(title)
        leg = plt.gca().get_legend()
        for txt in leg.get_texts():
            text = txt.get_text()
            txt.set_text(ptools.reduceSampleName(text,lifetime=True))
        plt.savefig(outdir+"dispEE_dxy.pdf")
        plt.close()
        
        plt.figure()
        h_chi2 = hsamp_qual.sum("vxy","dxy",overflow="all")
        plt.yscale("log")
        hist.plot1d(h_chi2,overlay="sample",order=samps_ct,density=True)
        plt.title(title)
        leg = plt.gca().get_legend()
        for txt in leg.get_texts():
            text = txt.get_text()
            txt.set_text(ptools.reduceSampleName(text,lifetime=True))
        plt.savefig(outdir+"dispEE_chi2.pdf")
        plt.close()

h_eff = histos["dispEE_eff"]
for pt in masses:
    for denom in denoms:
        m, dm = pt.split("-")[0], pt.split("-")[1]
        outdir = 'plots/ct-overlays/dispEE_eff/denom-{0}/{1}-{2}/'.format(denom,m,dm)
        os.makedirs(outdir,exist_ok=True)
        title = "$m_\chi = {0}$ GeV, $\Delta m_\chi = {1}$ GeV".format(m,dm)
        title += ", Denom: {0}".format(denom)
        samps_ct = [info[pt][ct] for ct in list(info[pt].keys())]
        
        plt.figure()
        hsamp_eff = h_eff.integrate("denom",int_range=denom)
        hist.plot1d(hsamp_eff,overlay="sample",order=samps_ct,density=True)
        plt.title(title)
        leg = plt.gca().get_legend()
        for txt in leg.get_texts():
            text = txt.get_text()
            txt.set_text(ptools.reduceSampleName(text,lifetime=True))
        plt.savefig(outdir+"dispEE_eff.pdf")
        plt.close()
```
