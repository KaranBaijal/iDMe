```python
import sys
import re
sys.path.append("/uscms/home/sbrightt/nobackup/iDM/iDMe_analysis/CMSSW_10_6_26/src/iDMe/python_analysis/analysisTools")
import plotTools as ptools
import matplotlib.pyplot as plt
import numpy as np
from coffea import hist
import importlib
import os
from analysisTools import fileAnalyzer, Analyzer

ptools.setDefaultStyle()
#plt.style.use(cfgDir+'plt_settings.mplstyle')
```

```python
histoLoc = "/uscms_data/d3/sbrightt/iDM/iDMe_analysis/CMSSW_10_6_26/src/iDMe/python_analysis/outputs/signal_2018_aEM"
histos = ptools.loadHistoFiles(histoLoc)
info, masses, cts = ptools.getSampleInfo(histos)
```

```python
# Plotting info about nearest reco vertex to the gen vertex in the reg-reg, low-reg, low-low categories
# (a) distance to vertex
# (b) vertex chi2
# (c) dR between associated electrons
# (d) dielectron mass
# (e) dielectron energy
# (f) vertex vxy
# (g) vertex vxy significance

reco_vtx_genMatch = histos['vtx_genMatchByDr']
reco_vtx_disp_genMatch = histos['vtx_disp_genMatchByDr']
reco_vtx_kin_genMatch = histos['vtx_kin_genMatchByDr']

for pt in list(info.keys()):
    for ct in list(info[pt].keys()):
        m, dm = pt.split("-")[0], pt.split("-")[1]
        samp = info[pt][ct]
        outdir = 'plots/{0}-{1}/vtx_nearestToGen_byType/ct-{2}/'.format(m,dm,ct)
        os.makedirs(outdir,exist_ok=True)
        title = r"Nearest to Gen Vtx, $m_\chi = {0}$, $\Delta m_\chi = {1}$, $c\tau = {2}$".format(m,dm,ct)

        # plotting distance from gen to reconstructed vertex
        plt.figure(figsize=(10,8))
        h_dist = reco_vtx_genMatch.integrate('sample',int_range=samp).sum('chi2','dR',overflow='all')
        plt.yscale('log')
        hist.plot1d(h_dist,overlay="type",density=True)
        plt.xlim([0,50])
        plt.xticks(np.arange(0,51,5))
        plt.title(title)
        plt.savefig(outdir+"distGenVtx_nearestToGen_byVtxType.pdf")
        plt.close()

        # plotting vertex reduced chi2
        plt.figure(figsize=(10,8))
        h_chi2 = reco_vtx_genMatch.integrate('sample',int_range=samp).sum('dist','dR',overflow='all')
        plt.yscale('log')
        hist.plot1d(h_chi2,overlay="type",density=True)
        plt.xlim([0,20])
        plt.xticks(np.arange(0,21,2))
        plt.title(title)
        plt.savefig(outdir+"chi2_nearestToGen_byVtxType.pdf")
        plt.close()

        # plotting electron dR
        plt.figure(figsize=(10,8))
        h_dR = reco_vtx_genMatch.integrate('sample',int_range=samp).sum('dist','chi2',overflow='all').rebin("dR",2)
        #plt.yscale('log')
        hist.plot1d(h_dR,overlay="type",density=True)
        plt.title(title)
        plt.savefig(outdir+"eeDR_nearestToGen_byVtxType.pdf")
        plt.close()
        
        # plotting vertex vxy
        plt.figure(figsize=(10,8))
        h_vxy = reco_vtx_disp_genMatch.integrate('sample',int_range=samp).sum('vxy_signif',overflow='all').rebin('vxy',2)
        plt.yscale('log')
        hist.plot1d(h_vxy,overlay="type",density=True)
        plt.title(title)
        plt.savefig(outdir+"ee_vxy_nearestToGen_byVtxType.pdf")
        plt.close()
        
        # plotting vertex vxy significance
        plt.figure(figsize=(10,8))
        h_vxy_signif = reco_vtx_disp_genMatch.integrate('sample',int_range=samp).sum('vxy',overflow='all')
        hist.plot1d(h_vxy_signif,overlay="type",density=True)
        plt.title(title)
        plt.savefig(outdir+"ee_vxySignif_nearestToGen_byVtxType.pdf")
        plt.close()
        
        # plotting dielectron mass
        plt.figure(figsize=(10,8))
        h_mee = reco_vtx_kin_genMatch.integrate('sample',int_range=samp).sum('energy',overflow='all')
        hist.plot1d(h_mee,overlay="type",density=True)
        plt.title(title)
        plt.savefig(outdir+"ee_mass_nearestToGen_byVtxType.pdf")
        plt.close()
        
        # plotting dielectron energy
        plt.figure(figsize=(10,8))
        h_mee = reco_vtx_kin_genMatch.integrate('sample',int_range=samp).sum('mass',overflow='all').rebin('energy',2)
        hist.plot1d(h_mee,overlay="type",density=True)
        plt.title(title)
        plt.savefig(outdir+"ee_energy_nearestToGen_byVtxType.pdf")
        plt.close()
```

```python
# Plotting match type for the globally nearest vertex in the event
# 0 = no vertices in event, 1 = reg-reg, 2 = low-reg, 3 = low-low
h_matchType_nearest = histos["vtx_nearestMatchTypeByDr"]

for pt in list(info.keys()):    
    m, dm = pt.split("-")[0], pt.split("-")[1]
    outdir = 'plots/{0}-{1}/vtx_nearestToGen_global/'.format(m,dm)
    os.makedirs(outdir,exist_ok=True)
    
    # overlay all lifetimes for a given mass point
    plt.figure()
    samps_ct = [info[pt][ct] for ct in list(info[pt].keys())]
    title = r"Global Nearest to Gen Vtx, $m_\chi = {0}$, $\Delta m_\chi = {1}$".format(m,dm)
    
    hist.plot1d(h_matchType_nearest,overlay="sample",order=samps_ct,density=True)
    plt.xticks(ticks=np.arange(0.5,4.5,step=1),labels=["No Vertices","Reg-Reg","Low-Reg","Low-Low"],fontsize=14)
    plt.ylim([0,1])
    plt.ylabel("Fraction of events",fontsize=14)
    plt.yticks(fontsize=14)
    plt.xlabel("")
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_type.pdf")
    plt.close()
```

```python
# Plotting vertex information for globally nearest-to-gen vertex
h_nearest_vtx = histos["vtx_nearestMatchByDr"]
h_nearest_vtx_kin = histos['vtx_kin_nearestMatchByDr']
h_nearest_vtx_disp = histos['vtx_disp_nearestMatchByDr']

for pt in list(info.keys()):    
    m, dm = pt.split("-")[0], pt.split("-")[1]
    outdir = 'plots/{0}-{1}/vtx_nearestToGen_global/'.format(m,dm)
    os.makedirs(outdir,exist_ok=True)
    samps_ct = [info[pt][ct] for ct in list(info[pt].keys())]
    title = r"Global Nearest to Gen Vtx, $m_\chi = {0}$, $\Delta m_\chi = {1}$".format(m,dm)
    
    # vertex type (where there is a vertex)
    plt.figure()
    h = h_nearest_vtx.sum("dist","chi2","dR",overflow="all")
    hist.plot1d(h,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    plt.xticks(ticks=np.arange(1.5,4.5,step=1),labels=["Reg-Reg","Low-Reg","Low-Low"],fontsize=14)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_type_onlyEvtsWithVtx.pdf")
    plt.close()
    
    # distance to gen vertex
    plt.figure()
    h = h_nearest_vtx.sum("type","chi2","dR",overflow="all")
    hist.plot1d(h,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_distToGen_onlyEvtsWithVtx.pdf")
    plt.close()
    
    # electron dR of vertex
    plt.figure()
    h = h_nearest_vtx.sum("type","chi2","dist",overflow="all")
    hist.plot1d(h,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_electronDR_onlyEvtsWithVtx.pdf")
    plt.close()
    
    # vertex chi2
    plt.figure()
    plt.yscale('log')
    h = h_nearest_vtx.sum("type","dR","dist",overflow="all")
    hist.plot1d(h,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_chi2_onlyEvtsWithVtx.pdf")
    plt.close()
    
    # plotting vertex vxy
    plt.figure(figsize=(10,8))
    h_vxy = h_nearest_vtx_disp.sum('vxy_signif','type',overflow='all').rebin('vxy',2)
    plt.yscale('log')
    hist.plot1d(h_vxy,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_vxy_onlyEvtsWithVtx.pdf")
    plt.close()

    # plotting vertex vxy significance
    plt.figure(figsize=(10,8))
    h_vxy_signif = h_nearest_vtx_disp.sum('vxy','type',overflow='all')
    hist.plot1d(h_vxy_signif,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_vxySignif_onlyEvtsWithVtx.pdf")
    plt.close()

    # plotting dielectron mass
    plt.figure(figsize=(10,8))
    h_mee = h_nearest_vtx_kin.sum('energy','type',overflow='all')
    hist.plot1d(h_mee,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_mass_onlyEvtsWithVtx.pdf")
    plt.close()

    # plotting dielectron energy
    plt.figure(figsize=(10,8))
    h_mee = h_nearest_vtx_kin.sum('mass','type',overflow='all').rebin('energy',2)
    hist.plot1d(h_mee,overlay="sample",order=samps_ct,density=True)
    plt.title(title)
    legend = plt.gca().get_legend()
    for txt in legend.get_texts():
        text = txt.get_text()
        txt.set_text(ptools.reduceSampleName(text,lifetime=True))
    plt.savefig(outdir+"nearest_vtx_energy_onlyEvtsWithVtx.pdf")
    plt.close()
```
